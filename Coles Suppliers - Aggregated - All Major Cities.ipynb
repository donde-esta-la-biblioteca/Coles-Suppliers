### Create DataFrames with desired columns

```python
import pandas as pd
```

```python
df_syd = pd.read_csv('sauce/Coles Data 2021-02-13 alexandria.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_perth = pd.read_csv('sauce/Coles Data 2021-02-13 dianella.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_brisbane = pd.read_csv('sauce/Coles Data 2021-02-13 fairfield.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_melb = pd.read_csv('sauce/Coles Data 2021-02-13 richmond-south.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_alicesprings = pd.read_csv('sauce/Coles Data 2021-02-14 alice-springs.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_goldcoast = pd.read_csv('sauce/Coles Data 2021-02-14 benowa-village.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_broome = pd.read_csv('sauce/Coles Data 2021-02-14 chinatown.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_adelaide = pd.read_csv('sauce/Coles Data 2021-02-14 churchill-centre.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_hobart = pd.read_csv('sauce/Coles Data 2021-02-14 glenorchy.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]
df_karratha = pd.read_csv('sauce/Coles Data 2021-02-14 karratha.csv')[['SKU', 'Brand', 'Product Name', 'Suppliers', 'Category']]

pd.set_option('display.max_rows', 200)
pd.set_option('display.max_colwidth', 200)
```

```python
df_syd.head()
```

### Concatenating all the DataFrames, remove duplicates and whatever else is needed

```python
df_agg = pd.concat([df_syd, df_perth, df_brisbane, 
                     df_melb, df_alicesprings, df_goldcoast, 
                     df_broome, df_adelaide, df_hobart, df_karratha], axis=0)

df_agg = df_agg.drop_duplicates('SKU')
df_agg.sort_values('SKU')
```

```python
df_agg = df_agg[df_agg['SKU'] != 'Page Unavailable'].reset_index(drop=True)
df_agg
```

### Clearing up the Suppliers column - remove square brackets and quotes

```python
df_agg['Suppliers'] = df_agg['Suppliers'].fillna('Suppliers Not Listed')
```

```python
import re
df_agg['Suppliers'] = df_agg['Suppliers'].apply(lambda x: x.strip("[ ]")).apply(lambda x: re.sub("'|"'"', "", x))
```

```python
df_agg['Suppliers']
```

```python
df_agg
```

### Explode!

```python
df_exploded_suppliers = df_agg.assign(Suppliers = df_agg['Suppliers'].str.split(', ')).explode('Suppliers')
df_exploded_suppliers.reset_index(drop=True, inplace=True)
df_exploded_suppliers
```

```python
df_exploded_suppliers.value_counts('Suppliers', ascending=False).nlargest(200)
```

```python
df_supplier_count = df_exploded_suppliers.value_counts('Suppliers', ascending=False)

df_supplier_count.to_excel('Supplier Count.xlsx')
```

```python
df_exploded_suppliers.groupby('Category')['Suppliers'].value_counts().head(200)
```

```python
df_suppliers_by_category = pd.DataFrame(df_exploded_suppliers.groupby('Category')['Suppliers'].value_counts())

df_suppliers_by_category.to_excel('Suppliers by Category.xlsx')
```
